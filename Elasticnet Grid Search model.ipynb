```python
#Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Step 1: Load the dataset 

```python
#Loading the required Dataset
df = pd.read_csv('car_price_dataset (1).csv')
```

```python
df.head()
```

### Step 2: Splitting Data into X and y features

```python
X = df.drop('price in $', axis = 1)
y = df['price in $']
```

```python
X.head(5)
```

```python
y.head(5)
```

### Step 3: Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 46)
```

### Step 4: Scaling the Features

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Step 5 : Modelling

```python
from sklearn.linear_model import ElasticNet
```

```python
model = ElasticNet()
```

```python
parameters = {"alpha": [0.01, 0.1, 0.5, 1, 5, 10, 20],"l1_ratio": [0.1, 0.3, 0.5, 0.7, 0.9, 1]}
```

### Step 6 : Create the Grid Search Model for the ElasticNet

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV( estimator=model, param_grid=parameters, scoring="neg_mean_squared_error", cv=10, verbose=1)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
# The best combination parameters
grid.best_estimator_
```

### Step 8: evaluate the Model

```python
# Grid Model Making Predictions
y_pred = grid.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Mean Square Error of Model
mse = mean_squared_error(y_test, y_pred)
```

```python
# Root mean Square Error of the Model
rmse = np.sqrt(mse)
rmse
```

```python
#R2_score of Model
r2_score(y_test, y_pred)
```

The average error in predictions is about $3617.The R² value of 0.81 which means that the model is able to explain 81% of the variation in car prices.
